```python
import cv2
from google.colab.patches import cv2_imshow  # Para trabajar en colab

face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades +
                                   "haarcascade_frontalface_default.xml") 
                  #Notar el cambio : agregar cv2.data.haarcascades al xml
img = cv2.imread("lafoto6.png")
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face_cascade.detectMultiScale(gray_img, scaleFactor=1.05,minNeighbors=5)
for x, y, w, h in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),10)
resized=cv2.resize(img,(int(img.shape[1]/3), int(img.shape[0]/3))) #Para version
                                                                   #OpenCV  3.1
cv2_imshow(resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
